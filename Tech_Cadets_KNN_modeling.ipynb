```python
import nltk
nltk.download('gutenberg')
from nltk.corpus import gutenberg
bible = gutenberg.raw('bible-kjv.txt')

# split all of the books of the bible into an array
books = bible.split('\n\n\n\n\n')

# index into John
John = books[53]

# split all of the verses which now become a list
John = John.split('\n\n')


#data cleaning



def clean_john(John):
  new_john = []
  for verse in John:
    # remove numbers and colons
    verse = ''.join([char for char in verse if not char.isdigit() and char != ':'])
    # replace "\n" with a space
    verse = verse.replace("\n", " ")
    new_john.append(verse)
  return new_john

John = clean_john(John)

John
```

```python

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
John.pop(0)
John

# Create a TF-IDF vectorizer
vectorizer = TfidfVectorizer(stop_words='english')

# # Fit the vectorizer to the John data
X = vectorizer.fit_transform(John)

# # Create a KMeans model with k=3 clusters
km = KMeans(n_clusters=12)

# # Fit the KMeans model to the TF-IDF vectors
km.fit(X)

# # Get the cluster labels for each sentence
labels = km.labels_

# # Print the cluster labels
for i, label in enumerate(labels):
  print(f"Sentence {i+1}: {John[i]}")
  print(f"Cluster: {label}")




```
